# **The angle between the vectors (1, 0, −1, 3) and (1, √3, 3, −3) in ℝ4**

```python
import numpy as np
v1=np.array([1,0,-1,3])
v2=np.array([1,np.sqrt(3),3,-3])

#claculate dot product
dot_product=np.dot(v1,v2)

#claculate norm(L2)(using Ecludian Distance)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v2)

cos_theta=dot_product/(norm_v1*norm_v2)
theta=np.arccos(cos_theta)

print(f"dot product of V1 and V2= {dot_product}")
print(f"norm of V1= {norm_v1}")
print(f"norm of V2= {norm_v2}")
print(f"cos theta= {cos_theta}")
print(f"angle between V1 and V2= {theta}")
```

#  Which of the angles (if any) of triangle ABC, with A = (1, −2, 0), B = (2, 1, −2), and C = (6, −1, −3), is a right angle? Answer: the angle at vertex

```python
A=np.array([1,-2,0])
B=np.array([2,1,-2])
C=np.array([6,-1,-3])

AB=B-A
AC=C-A
dot_pr_AB=np.dot(AB,AC)
print(f"dot product of AB and AC= {dot_pr_AB}")

BA=A-B
BC=C-B
dot_pr_BA=np.dot(BA,BC)
print(f"dot product of BA and BC= {dot_pr_BA}")

CA=A-C
CB=B-C
dot_pr_CA=np.dot(CA,CB)
print(f"dot product of CA and CB= {dot_pr_CA}")

if dot_pr_AB==0:
  print("Angle A is right angle")

if dot_pr_BA==0:
  print("Angle B is right angle")

if dot_pr_CA==0:
  print("Angle C is right angle")
```
