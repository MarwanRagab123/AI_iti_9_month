## Write a function that accepts two arguments (length, start) to generate an array of a specific length filled with integer numbers increased by one from start.

```python
def arr(length,start):
    array = []
    for i in range(length):
        array.append(start + i)
    return array
print(arr(5, 10))
```

# write a function that takes a number as an argument and if the
# number divisible by 3 return "Fizz" and if it is divisible by 5 return
# "buzz" and if is is divisible by both return " FizzBuzz

```python
def fun(num):
    if num%3==0:
        return "Fizz"
    elif num%5==0:
        return "Buzz"
    elif num%3==0 and num%5==0:
        return "FizzBuzz"
    else:
        return num
print(fun(15))
```

```python
# Word guessing game (hangman)
# ○
# A list of words will be hardcoded in your program, out of
# which the interpreter will
# ○
# choose 1 random word.
# ○
# The user first must input their names
# ○
# Ask the user to guess any alphabet. If the random word
# contains that alphabet, it
# ○
# will be shown as the output(with correct placement)
# ○
# Else the program will ask you to guess another alphabet.
# ○
# Give 7 turns maximum to guess the complete word.

import random
def hangman():
    word_list = ['python', 'java', 'kotlin', 'javascript']
    chosen_word = random.choice(word_list)
    guessed_letters = set()
    attempts = 7
    display_word = ['_'] * len(chosen_word)

    print("Welcome to Hangman!")
    name = input("Enter your name: ")
    print(f"Hello, {name}! Let's start the game.")

    while attempts > 0 and '_' in display_word:
        print("\n" + " ".join(display_word))
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single alphabetic character.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        guessed_letters.add(guess)

        if guess in chosen_word:
            for index, letter in enumerate(chosen_word):
                if letter == guess:
                    display_word[index] = guess
        else:
            attempts -= 1
            print(f"Wrong guess! You have {attempts} attempts left.")

    if '_' not in display_word:
        print(f"\nCongratulations, {name}! You guessed the word: {chosen_word}")
    else:
        print(f"\nSorry, {name}. You've run out of attempts. The word was: {chosen_word}")
hangman()
```
